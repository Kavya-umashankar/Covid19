#### Data Cleaning

```python
# total rows 392,removing all columns which had nan or null values sum >= 325
lstofunwantedcols = [col for col in NJ.columns if NJ[col].isnull().sum()>=325]
```

```python
lstofunwantedcols
```

```python
len(lstofunwantedcols)
```

```python
#removing all the Unwanted Columns 
NJ.drop(lstofunwantedcols, axis=1, inplace=True)
```

```python
NJ.head()
```

```python
#date and state column, Categorical columns
cat_name = [fea for fea in NJ.columns if NJ[fea].dtypes =='object']
```

```python
cat_col = NJ[cat_name]
num_col= NJ[NJ.columns.difference(cat_name)]
```
